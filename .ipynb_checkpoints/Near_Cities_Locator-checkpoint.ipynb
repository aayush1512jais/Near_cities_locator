# Data Pre-Processing

```python
city_df=city_df.fillna(0)
master_df=master_df.fillna(0)
cityList=list(city_df.columns.values)


#print("List of cities: ",cityList)
```

```python
def get_nearCities(cityName,cityList):
    nearCities = master_df.loc[cityName][cityList].reindex()
    nearCities = nearCities.sort_values()
    #nearsCities = sorted(nearCities,key=lambda x:x[1])
    return nearCities
```

```python
nearCities = get_nearCities(consumer_city,cityList).dropna()
nearCities_df=pd.DataFrame({"Cities":nearCities.index,"Distance (in Km)":nearCities.values,"Time(in Hours)": nearCities.values/55})
nearCities_df
```
